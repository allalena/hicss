```python
import os
import time
import random
import copy
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# For CPU memory usage
import psutil
from PIL import Image
from tqdm import tqdm
from collections import Counter

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, Subset
from torchvision import transforms, models

# If you have a GPU, you can keep high precision:
torch.set_float32_matmul_precision('high')

# For monitoring GPU memory usage
try:
    import pynvml
    pynvml.nvmlInit()
    gpu_monitor_available = True
except ImportError:
    print("pynvml not installed. GPU memory monitoring disabled.")
    gpu_monitor_available = False

from sklearn.metrics import classification_report, confusion_matrix
import warnings
warnings.filterwarnings(
    "ignore",
    category=FutureWarning,
    message=r".*torch\.cuda\.amp.*"
)

# Metric learning library
try:
    from pytorch_metric_learning import miners, losses
    metric_learning_available = True
except ImportError:
    print("pytorch-metric-learning not installed. Some metric features unavailable.")
    metric_learning_available = False

# EasyFSL for few-shot
try:
    from easyfsl.samplers import TaskSampler
    easyfsl_available = True
except ImportError:
    print("easyfsl not installed. Few-shot evaluation unavailable.")
    easyfsl_available = False

################################################################################
# 1) Reproducibility
################################################################################
seed = 42
random.seed(seed)
np.random.seed(seed)
torch.manual_seed(seed)
torch.cuda.manual_seed_all(seed)

torch.backends.cudnn.deterministic = True
torch.backends.cudnn.benchmark = False

try:
    from google.colab import drive
    drive.mount('/content/drive')
except:
    print("Not on Colab or google.colab not available.")

################################################################################
# 2) Custom Dataset & Subset Classes
################################################################################
class WeedNet(Dataset):
    """
    Expects:
      root/
        class_0/
          imgA.jpg
          ...
        class_1/
          ...
    """
    def __init__(self, root, transform=None):
        self.root = root
        self.transform = transform
        self.classes = sorted([d for d in os.listdir(root) if os.path.isdir(os.path.join(root,d))])
        self.class_to_idx = {cls_name: i for i, cls_name in enumerate(self.classes)}
        self.samples = []
        for cls_name in self.classes:
            folder = os.path.join(root, cls_name)
            for fname in os.listdir(folder):
                if fname.lower().endswith(('.jpg','.jpeg','.png')):
                    path = os.path.join(folder,fname)
                    label= self.class_to_idx[cls_name]
                    self.samples.append((path,label))

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        path, lab = self.samples[idx]
        image = Image.open(path).convert("RGB")
        if self.transform:
            image = self.transform(image)
        return image, lab

    def get_labels(self):
        return [s[1] for s in self.samples]

    def get_classes(self):
        return self.classes

    def get_class_to_idx(self):
        return self.class_to_idx


class SubsetWithLabels(Dataset):
    def __init__(self, subset: Subset):
        self.subset = subset
        self.indices = subset.indices
        original_dataset = subset.dataset

        if hasattr(original_dataset, "get_labels") and hasattr(original_dataset,"get_classes"):
            all_labels = original_dataset.get_labels()
            self._all_classes = original_dataset.get_classes()
            self._all_class_to_idx = original_dataset.get_class_to_idx()

            self._subset_labels = [all_labels[i] for i in self.indices]
            self._subset_labels = [int(lbl) for lbl in self._subset_labels]
            unique_labels_in_subset = sorted(list(set(self._subset_labels)))

            self._subset_classes = [self._all_classes[i] for i in unique_labels_in_subset]
            self._subset_class_to_idx = {
                cls_name: i for i, cls_name in enumerate(self._subset_classes)
            }
        else:
            # fallback
            self._subset_labels = []
            labels=[]
            for i in tqdm(self.indices, desc="Inferring labels"):
                _, lab= original_dataset[i]
                labels.append(lab)
            self._subset_labels= [int(lab) for lab in labels]
            unique_labels_in_subset= sorted(list(set(self._subset_labels)))
            self._subset_classes= [f"Class_{u}" for u in unique_labels_in_subset]
            self._subset_class_to_idx= {nm:i for i,nm in enumerate(self._subset_classes)}

    def __len__(self):
        return len(self.indices)

    def __getitem__(self, idx):
        return self.subset[idx]

    def get_labels(self):
        return self._subset_labels

class SubsetWithTransform(Dataset):
    def __init__(self, subset: SubsetWithLabels, transform=None):
        self.subset = subset
        self.transform = transform

    def __len__(self):
        return len(self.subset)

    def __getitem__(self, idx):
        image, label = self.subset[idx]
        if self.transform:
            image= self.transform(image)
        return image, label

    def get_labels(self):
        return self.subset.get_labels()

################################################################################
# 3) Hyperparameters & Setup
################################################################################
N_WAY=5
N_SHOT=5
N_QUERY=15
N_VAL_EPISODES=1000
N_EVALUATION_TASKS=1000
N_TRAINING_EPOCHS=25
BATCH_SIZE_FT=32
LEARNING_RATE_FT=1e-4
WEIGHT_DECAY_FT=1e-5
PATIENCE_FT=5

DISTILL_ALPHA=1.0
TRIPLET_BETA=1.0
TRIPLET_MARGIN=1.0
DISTILL_EPOCHS=50
DISTILL_LR=1e-4
DISTILL_PATIENCE=5
DISTILL_EVAL_FREQ=5

P_CLASSES=8
K_SAMPLES=4
image_size=224

root_dir = "/content/drive/MyDrive/WeedClassification/Datasets/WeedNet"
model_save_dir= "/content/drive/MyDrive/WeedClassification/saved_models/"
os.makedirs(model_save_dir, exist_ok=True)

################################################################################
# 4) Transforms
################################################################################
train_transform= transforms.Compose([
    transforms.RandomResizedCrop(image_size, scale=(0.8,1.0)),
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomVerticalFlip(p=0.5),
    transforms.RandomRotation(20),
    transforms.ColorJitter(brightness=0.3, contrast=0.3, saturation=0.2, hue=0.1),
    transforms.ToTensor(),
    transforms.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225])
])
val_transform= transforms.Compose([
    transforms.Resize(int(image_size*1.15)),
    transforms.CenterCrop(image_size),
    transforms.ToTensor(),
    transforms.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225])
])
test_transform= transforms.Compose([
    transforms.Resize(int(image_size*1.15)),
    transforms.CenterCrop(image_size),
    transforms.ToTensor(),
    transforms.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225])
])

################################################################################
# 5) Split Dataset (Train/Val/Test)
################################################################################
full_dataset= WeedNet(root_dir, transform=None)
all_labels= full_dataset.get_labels()
num_classes_total= len(set(all_labels))

from sklearn.model_selection import StratifiedShuffleSplit
splitter= StratifiedShuffleSplit(n_splits=1, test_size=0.30, random_state=seed)
train_idx, valtest_idx= next(splitter.split(range(len(full_dataset)), all_labels))

valtest_labels= [all_labels[i] for i in valtest_idx]
splitter2= StratifiedShuffleSplit(n_splits=1, test_size=0.5, random_state=seed)
val_idx_part, test_idx_part= next(splitter2.split(valtest_idx, valtest_labels))

train_sub_no_labels= Subset(full_dataset, train_idx)
val_sub_no_labels  = Subset(full_dataset, [valtest_idx[i] for i in val_idx_part])
test_sub_no_labels = Subset(full_dataset, [valtest_idx[i] for i in test_idx_part])

train_labeled= SubsetWithLabels(train_sub_no_labels)
val_labeled  = SubsetWithLabels(val_sub_no_labels)
test_labeled = SubsetWithLabels(test_sub_no_labels)

train_dataset= SubsetWithTransform(train_labeled, transform=train_transform)
val_dataset  = SubsetWithTransform(val_labeled,   transform=val_transform)
test_dataset = SubsetWithTransform(test_labeled,  transform=test_transform)

print(f"Train samples: {len(train_dataset)}, Val samples: {len(val_dataset)}, Test samples: {len(test_dataset)}")
print(f"Number of Classes (Total): {num_classes_total}")

################################################################################
# 6) Teacher Model (EffNet-B7) but NO partial fine-tuning
################################################################################
# We load EfficientNet-B7 pretrained and freeze everything,
# so effectively "partial fine-tuning" is disabled.
from torchvision.models import efficientnet_b7, EfficientNet_B7_Weights

teacher_model= efficientnet_b7(weights=EfficientNet_B7_Weights.IMAGENET1K_V1)
# Replace final layer => 2048-d embeddings
teacher_model.classifier= nn.Sequential(
    nn.Dropout(p=0.4, inplace=True),
    nn.Linear(in_features=2560, out_features=2048)
)

# Freeze all parameters => NO fine-tuning
for param in teacher_model.parameters():
    param.requires_grad= False

teacher_model= teacher_model.cuda()
try:
    if torch.__version__>= "2.0.0":
        torch._dynamo.config.force_parameter_static_shapes= False
        teacher_model= torch.compile(teacher_model)
except:
    pass

# We'll call this teacher_extractor
def teacher_extractor(x):
    """Return 2048-d embedding from the frozen teacher."""
    return teacher_model(x)

################################################################################
# 7) Student: Weighted Ensemble of 3 Light Models
################################################################################
from torchvision.models import (
    mobilenet_v3_small, MobileNet_V3_Small_Weights,
    shufflenet_v2_x1_0, ShuffleNet_V2_X1_0_Weights,
    efficientnet_b0, EfficientNet_B0_Weights
)

class StudentEmbeddingNet(nn.Module):
    def __init__(self, backbone, in_features, dropout_p=0.5):
        super().__init__()
        self.backbone= backbone
        self.pool= nn.AdaptiveAvgPool2d((1,1))
        self.flatten= nn.Flatten()
        self.dropout= nn.Dropout(p=dropout_p)
        self.project= nn.Linear(in_features, 2048)

    def forward(self,x):
        x= self.backbone(x)
        x= self.pool(x)
        x= self.flatten(x)
        x= self.dropout(x)
        x= self.project(x)
        return x

# 1) MobileNetV3-Small
mobv3= mobilenet_v3_small(weights=MobileNet_V3_Small_Weights.IMAGENET1K_V1)
student_1= StudentEmbeddingNet(
    backbone= mobv3.features,
    in_features=576,
    dropout_p=0.5
).cuda()

# 2) ShuffleNetV2
shufflenet= shufflenet_v2_x1_0(weights=ShuffleNet_V2_X1_0_Weights.IMAGENET1K_V1)
shufflenet_features= nn.Sequential(*list(shufflenet.children())[:-1])
student_2= StudentEmbeddingNet(
    backbone= shufflenet_features,
    in_features=1024,
    dropout_p=0.5
).cuda()

# 3) EfficientNet-B0
effb0= efficientnet_b0(weights=EfficientNet_B0_Weights.IMAGENET1K_V1)
effb0_backbone= effb0.features
student_3= StudentEmbeddingNet(
    backbone= effb0_backbone,
    in_features=1280,
    dropout_p=0.5
).cuda()

try:
    if torch.__version__>= "2.0.0":
        torch._dynamo.config.force_parameter_static_shapes= False
        student_1= torch.compile(student_1)
        student_2= torch.compile(student_2)
        student_3= torch.compile(student_3)
except:
    pass

class WeightedEnsembleEmbeddingNet(nn.Module):
    """
    Weighted sum of embeddings from multiple students.
    """
    def __init__(self, students):
        super().__init__()
        self.students= nn.ModuleList(students)
        self.weights= nn.Parameter(torch.ones(len(students)), requires_grad=True)

    def forward(self, x):
        embs= [st(x) for st in self.students]
        alpha= torch.softmax(self.weights, dim=0)
        out= 0
        for i, e in enumerate(embs):
            out+= alpha[i]* e.float()
        return out

class SiameseTripletNetwork(nn.Module):
    def __init__(self, backbone):
        super().__init__()
        self.backbone= backbone
    def forward(self,x):
        return self.backbone(x)

students_ensemble= WeightedEnsembleEmbeddingNet([student_1, student_2, student_3])
fewshot_model= SiameseTripletNetwork(students_ensemble).cuda()

################################################################################
# 8) Distillation + Triplet Training
################################################################################
def evaluate_few_shot_prototypical(model_, dataset_, n_way, n_shot, n_query, n_tasks):
    """
    Prototypical N-way, N-shot evaluation if easyfsl is installed.
    """
    if not easyfsl_available:
        print("easyfsl not installed => skipping few-shot.")
        return 0.0
    model_.eval()
    sampler= TaskSampler(dataset_, n_way=n_way, n_shot=n_shot, n_query=n_query, n_tasks=n_tasks)
    loader= DataLoader(
        dataset_,
        batch_sampler= sampler,
        collate_fn= sampler.episodic_collate_fn,
        num_workers=4
    )
    total_correct=0
    total_count=0

    with torch.no_grad():
        for (support_imgs, support_lbls, query_imgs, query_lbls, _) in loader:
            support_imgs= support_imgs.cuda()
            query_imgs= query_imgs.cuda()
            support_lbls= support_lbls.cpu()
            query_lbls= query_lbls.cpu()

            s_emb= model_(support_imgs).cpu()
            q_emb= model_(query_imgs).cpu()

            unique_labels= torch.unique(support_lbls)
            centroids=[]
            for ulab in unique_labels:
                idx= (support_lbls==ulab)
                centroids.append(s_emb[idx].mean(dim=0))
            centroids= torch.stack(centroids, dim=0)

            dists= torch.cdist(q_emb, centroids)
            preds_idx= torch.argmin(dists, dim=1)
            mapped_preds= [unique_labels[p].item() for p in preds_idx]

            correct=0
            for i, mp in enumerate(mapped_preds):
                if mp== query_lbls[i].item():
                    correct+=1
            total_correct+= correct
            total_count+= len(query_lbls)

    acc= 100.0* total_correct/ total_count if total_count else 0.0
    return acc

def train_combined_distillation(
    model_,
    teacher_extractor,
    train_data,
    val_data,
    p_classes,
    k_samples,
    alpha,
    beta,
    margin,
    epochs=30,
    lr=1e-4,
    weight_decay=1e-5,
    patience=5,
    eval_freq=5,
    test_data=None
):
    if not metric_learning_available or not easyfsl_available:
        print("Skipping => metric/easyfsl not available.")
        return [],[],[],[],[],[]

    train_loader= DataLoader(train_data, batch_size=64, shuffle=True, num_workers=4)
    criterion_mse= nn.MSELoss()
    from pytorch_metric_learning import miners, losses

    semi_hard_miner= miners.TripletMarginMiner(margin=margin, type_of_triplets="semihard")
    miner_triplet_loss= losses.TripletMarginLoss(margin=margin)

    optimizer= optim.Adam(model_.parameters(), lr=lr, weight_decay=weight_decay)
    scheduler= optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=epochs, eta_min=1e-6)
    scaler= torch.cuda.amp.GradScaler()

    best_val_acc= -1
    patience_cnt= 0
    best_model_state= copy.deepcopy(model_.state_dict())

    train_total_loss_hist= []
    train_mse_loss_hist= []
    train_triplet_loss_hist= []
    val_acc_hist= []
    val_loss_hist= []
    eval_epoch_list= []
    train_fs_hist= []

    print(f"Distill+Triplet => alpha={alpha}, beta={beta}, margin={margin}...")

    for ep in range(1, epochs+1):
        model_.train()
        # teacher_extractor is a function => no grad
        running_loss_combined=0.0
        running_loss_mse=0.0
        running_loss_trip=0.0
        n_batch=0

        for imgs, lbls in tqdm(train_loader, desc=f"Epoch {ep}/{epochs} [semi-hard triplet]"):
            imgs, lbls= imgs.cuda(), lbls.cuda()

            with torch.no_grad():
                teacher_emb= teacher_extractor(imgs).detach()
            student_emb= model_(imgs)
            loss_mse= criterion_mse(student_emb, teacher_emb)

            # semi-hard triplet
            emb= model_(imgs)
            triplets= semi_hard_miner(emb, lbls)
            if len(triplets[0])>0:
                loss_trip= miner_triplet_loss(emb, lbls, triplets)
            else:
                loss_trip= torch.tensor(0.0, device=imgs.device)

            total_loss= alpha*loss_mse + beta*loss_trip

            optimizer.zero_grad()
            with torch.cuda.amp.autocast():
                pass
            scaler.scale(total_loss).backward()
            scaler.step(optimizer)
            scaler.update()

            running_loss_combined+= total_loss.item()
            running_loss_mse+= loss_mse.item()
            running_loss_trip+= loss_trip.item()
            n_batch+=1

        ep_combined= running_loss_combined/ n_batch if n_batch else 0.0
        ep_mse= running_loss_mse/ n_batch if n_batch else 0.0
        ep_trip= running_loss_trip/ n_batch if n_batch else 0.0

        train_total_loss_hist.append(ep_combined)
        train_mse_loss_hist.append(ep_mse)
        train_triplet_loss_hist.append(ep_trip)

        scheduler.step()

        # Evaluate few-shot on train data for logging
        model_.eval()
        train_fs_acc=0.0
        if ep%eval_freq==0 or ep==epochs:
            train_fs_acc= evaluate_few_shot_prototypical(model_, train_data, N_WAY, N_SHOT, N_QUERY, N_VAL_EPISODES)
        train_fs_hist.append(train_fs_acc)

        # Evaluate few-shot on val
        val_fs_acc=0.0
        if ep%eval_freq==0 or ep==epochs:
            print(f"[Distill+Triplet] EVAL epoch={ep}, semi-hard")
            if val_data and len(val_data)>0:
                val_fs_acc= evaluate_few_shot_prototypical(model_, val_data, N_WAY, N_SHOT, N_QUERY, N_VAL_EPISODES)
            val_acc_hist.append(val_fs_acc)
            eval_epoch_list.append(ep)

            val_loss_hist.append(0.0)
            print(f"Epoch {ep} => CombinedLoss={ep_combined:.4f}, TrainFS={train_fs_acc:.2f}%, ValFS={val_fs_acc:.2f}%")

            # Early stopping
            if val_fs_acc> best_val_acc:
                best_val_acc= val_fs_acc
                patience_cnt=0
                best_model_state= copy.deepcopy(model_.state_dict())
                print(f"  new best valFS={best_val_acc:.2f}%")
            else:
                patience_cnt+=1
                print(f"  no improvement => patience={patience_cnt}/{patience}")
                if patience_cnt>= patience:
                    print(f"Early stopping at epoch={ep}")
                    break

    model_.load_state_dict(best_model_state)
    print(f"Loaded best model => val FS={best_val_acc:.2f}%")

    if test_data is not None:
        test_fs_acc= evaluate_few_shot_prototypical(model_, test_data, N_WAY, N_SHOT, N_QUERY, N_EVALUATION_TASKS)
        print(f"[Distill+Triplet] Final Test FS acc={test_fs_acc:.2f}%")

    # Plot train loss
    if train_total_loss_hist:
        plt.figure(figsize=(10,6))
        plt.plot(range(1, len(train_total_loss_hist)+1), train_total_loss_hist, label="Total Combined Loss")
        plt.plot(range(1, len(train_mse_loss_hist)+1), train_mse_loss_hist, label="MSE Distillation", linestyle='--')
        plt.plot(range(1, len(train_triplet_loss_hist)+1), train_triplet_loss_hist, label="Triplet Loss", linestyle=':')
        plt.title("Distillation+Triplet => Train Loss Over Epochs")
        plt.xlabel("Epoch")
        plt.ylabel("Loss")
        plt.legend()
        plt.grid(True)
        plt.show()

    # Plot few-shot
    if train_fs_hist and val_acc_hist and eval_epoch_list:
        plt.figure(figsize=(10,6))
        offset= len(train_fs_hist)- len(eval_epoch_list)
        train_curve= train_fs_hist[offset:] if offset>=0 else train_fs_hist
        plt.plot(eval_epoch_list, train_curve, 'b-', label="Train FS Acc")
        plt.plot(eval_epoch_list, val_acc_hist, 'ro-', label="Val FS Acc")
        plt.title("Distill+Triplet => Train & Val FS Accuracy")
        plt.xlabel("Epoch")
        plt.ylabel("Accuracy (%)")
        plt.legend()
        plt.grid(True)
        plt.show()

    return (train_total_loss_hist, train_mse_loss_hist, train_triplet_loss_hist,
            val_acc_hist, val_loss_hist, eval_epoch_list)


# If libs are available, train
if metric_learning_available and easyfsl_available:
    (train_loss_hist,
     train_mse_hist,
     train_triplet_hist,
     val_acc_hist,
     val_loss_hist,
     eval_epochs)= train_combined_distillation(
        fewshot_model,
        teacher_extractor,     # frozen teacher
        train_dataset,
        val_dataset,
        p_classes=P_CLASSES,
        k_samples=K_SAMPLES,
        alpha=DISTILL_ALPHA,
        beta=TRIPLET_BETA,
        margin=TRIPLET_MARGIN,
        epochs=DISTILL_EPOCHS,
        lr=DISTILL_LR,
        weight_decay=1e-5,
        patience=DISTILL_PATIENCE,
        eval_freq=DISTILL_EVAL_FREQ,
        test_data=test_dataset
    )
else:
    print("Skipping => metric or easyfsl not installed.")

################################################################################
# 9) Multi-run N-Way N-Shot (use n_runs=3)
################################################################################
def multi_run_few_shot_eval(
    model_,
    dataset_,
    ways_list,
    shots_list,
    n_query,
    n_tasks,
    n_runs=3  # changed from 5 to 3
):
    if not easyfsl_available:
        return
    results= []
    for ways in ways_list:
        for shots in shots_list:
            print(f"\n=== Evaluate: {ways}-Way, {shots}-Shot ===")
            run_accs= []
            for r in range(n_runs):
                print(f"   Run {r+1}/{n_runs} ...")
                acc= evaluate_few_shot_prototypical(model_, dataset_, ways, shots, n_query, n_tasks)
                run_accs.append(acc)
                print(f"   Accuracy: {acc:.2f}%")
            mean_acc= np.mean(run_accs)
            std_acc= np.std(run_accs)
            print(f"--> {ways}-Way {shots}-Shot => mean={mean_acc:.2f}%, std={std_acc:.2f}%")
            results.append((ways, shots, mean_acc, std_acc))
    return results

if metric_learning_available and easyfsl_available:
    ways_to_try= [3, 5, 7]
    shots_to_try= [1, 5, 10]
    final_results= multi_run_few_shot_eval(
        fewshot_model,
        test_dataset,
        ways_list= ways_to_try,
        shots_list= shots_to_try,
        n_query= N_QUERY,
        n_tasks= N_EVALUATION_TASKS,
        n_runs= 3  # changed from 5 to 3
    )
    print("\n----- Final (N-Way, N-Shot) Results on Test -----")
    for (w, s, m, sd) in final_results:
        print(f"{w}-Way {s}-Shot => Mean={m:.2f}% +/- {sd:.2f}%")
else:
    print("Skipping final multi-run => missing easyfsl.")

################################################################################
# 10) Quantize for Edge
################################################################################
if metric_learning_available and easyfsl_available:
    save_path= os.path.join(model_save_dir, "fewshot_weednet_siamese_ensemble_frozen_teacher.pth")
    torch.save(fewshot_model.state_dict(), save_path)
    orig_size= os.path.getsize(save_path)
    print(f"Original model size: {orig_size/1e6:.2f} MB")

    # Dynamic quantization on CPU
    quant_model= copy.deepcopy(fewshot_model).cpu()
    quant_model.eval()
    try:
        quant_model= torch.quantization.quantize_dynamic(quant_model, {nn.Linear}, dtype=torch.qint8)
        quant_path= os.path.join(model_save_dir,"fewshot_weednet_siamese_ensemble_frozen_teacher_quant.pth")
        torch.save(quant_model.state_dict(), quant_path)
        q_size= os.path.getsize(quant_path)
        red= 100.0*(orig_size- q_size)/ orig_size
        print(f"Quant model size: {q_size/1e6:.2f} MB => reduced by {red:.2f}%")
    except Exception as e:
        print(f"Quant failed: {e}")
else:
    print("Skipping quant => missing easyfsl or no training done.")

################################################################################
# 11) Speed Demo
################################################################################
def measure_few_shot_inference_speed(model_, device='cuda', n_way=5, n_shot=5, n_query=5, num_tasks=10):
    if not easyfsl_available:
        print("easyfsl not available => skipping speed test.")
        return
    is_quant= any("Quantized" in str(type(m)) for m in model_.modules())
    if is_quant:
        device='cpu'
        print("Quantized => forcing CPU.")
    model_.eval().to(device)

    support= torch.randn(n_way*n_shot,3,image_size,image_size, device=device)
    query= torch.randn(n_way*n_query,3,image_size,image_size, device=device)
    support_labels= torch.arange(n_way).unsqueeze(1).expand(n_way,n_shot).reshape(-1).to(device)

    # warmup
    for _ in range(2):
        with torch.no_grad():
            s_emb= model_(support)
            q_emb= model_(query)
            cents=[]
            for w in range(n_way):
                idx= (support_labels==w)
                cents.append(s_emb[idx].mean(dim=0))
            cents= torch.stack(cents, dim=0)
            dists= torch.cdist(q_emb, cents)

    if device=='cuda':
        torch.cuda.synchronize()
    t0= time.time()

    for _ in range(num_tasks):
        with torch.no_grad():
            s_emb= model_(support)
            q_emb= model_(query)
            cents=[]
            for w in range(n_way):
                idx= (support_labels==w)
                cents.append(s_emb[idx].mean(dim=0))
            if not cents:
                break
            cents= torch.stack(cents, dim=0)
            _= torch.cdist(q_emb, cents)

    if device=='cuda':
        torch.cuda.synchronize()
    t1= time.time()
    total_time= t1- t0
    if num_tasks>0:
        avg_time= total_time/ num_tasks
        total_imgs= (n_way*n_shot)+(n_way*n_query)
        throughput= (total_imgs*num_tasks)/ total_time
        print(f"\n[Few-Shot Inference Speed => {n_way}-Way, {n_shot}-Shot, {n_query}-Query, #tasks={num_tasks}]")
        print(f"  avg time/task= {avg_time:.4f}s, throughput= {throughput:.2f} samples/s")

    # memory usage
    process= psutil.Process(os.getpid())
    mem_info= process.memory_info().rss/(1024**2)
    print(f"CPU Memory= {mem_info:.2f} MB")

    if gpu_monitor_available and device=='cuda':
        try:
            handle= pynvml.nvmlDeviceGetHandleByIndex(0)
            info= pynvml.nvmlDeviceGetMemoryInfo(handle)
            print(f"GPU Memory= {info.used/(1024**2):.2f} MB")
        except:
            pass

if metric_learning_available and easyfsl_available:
    measure_few_shot_inference_speed(fewshot_model, device='cuda', n_way=5, n_shot=5, n_query=5, num_tasks=10)
else:
    print("Skipping inference speed => missing easyfsl.")


print("\n===== Pipeline Complete (No Partial Fine-Tuning) =====\n"
      "1) Teacher (EffNet-B7) is entirely frozen.\n"
      "2) Distill + Triplet to ensemble student.\n"
      "3) Few-shot evaluation with n_runs=3.\n"
      "4) Optional quantization.\n"
      "5) Speed test.\n")
```
