```python
import os
import time
import random
import copy
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# For CPU memory usage
import psutil
from PIL import Image
from tqdm import tqdm
from collections import Counter

################################################################################
# IMPORTANT:
# Make sure no local file/folder is named "torch.py" or "torch" to avoid
# overshadowing the real PyTorch library.
################################################################################

# Actual torch imports:
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, Subset
from torchvision import transforms, models

# If you have a GPU, can keep higher precision.
torch.set_float32_matmul_precision('high')

# For GPU memory usage
try:
    import pynvml
    pynvml.nvmlInit()
    gpu_monitor_available = True
except ImportError:
    print("pynvml not installed. GPU memory monitoring disabled.")
    gpu_monitor_available = False

# For classification metrics
from sklearn.metrics import classification_report, confusion_matrix, f1_score, accuracy_score

import warnings
warnings.filterwarnings(
    "ignore",
    category=FutureWarning,
    message=r".*torch\.cuda\.amp.*"
)

# Metric learning library (used for semi-hard triplet miner, etc.)
try:
    from pytorch_metric_learning import miners, losses
    metric_learning_available = True
except ImportError:
    metric_learning_available = False
    print("pytorch-metric-learning not installed; skipping advanced miner usage.")

# EasyFSL for few-shot
try:
    from easyfsl.samplers import TaskSampler
    easyfsl_available = True
except ImportError:
    easyfsl_available = False
    print("easyfsl not installed; few-shot evaluation unavailable.")


###############################################################################
# 1) Reproducibility & Colab
###############################################################################
seed = 42
random.seed(seed)
np.random.seed(seed)
torch.manual_seed(seed)
torch.cuda.manual_seed_all(seed)

# For reproducibility
torch.backends.cudnn.deterministic = True
torch.backends.cudnn.benchmark = False

# If running on Google Colab + Drive
try:
    from google.colab import drive
    drive.mount('/content/drive')
except:
    print("Not on Colab or google.colab not available.")


###############################################################################
# 2) Custom WeedNet Dataset + Subset Helpers
###############################################################################
class WeedNet(Dataset):
    """
    Expects folder structure:
      root/
          class_0/
              imgA.jpg
              ...
          class_1/
              ...
    """
    def __init__(self, root, transform=None):
        self.root = root
        self.transform = transform
        self.classes = sorted(
            d for d in os.listdir(root) if os.path.isdir(os.path.join(root, d))
        )
        self.class_to_idx = {cls_name: i for i, cls_name in enumerate(self.classes)}
        self.samples = []
        for cls_name in self.classes:
            folder = os.path.join(root, cls_name)
            for fname in os.listdir(folder):
                if fname.lower().endswith(('.jpg', '.jpeg', '.png')):
                    path = os.path.join(folder, fname)
                    label = self.class_to_idx[cls_name]
                    self.samples.append((path, label))

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        p, lb = self.samples[idx]
        img = Image.open(p).convert("RGB")
        if self.transform:
            img = self.transform(img)
        return img, lb

    def get_labels(self):
        return [s[1] for s in self.samples]

    def get_classes(self):
        return self.classes

    def get_class_to_idx(self):
        return self.class_to_idx


class SubsetWithLabels(Dataset):
    def __init__(self, subset: Subset):
        self.subset = subset
        self.indices = subset.indices
        original_dataset = subset.dataset

        if hasattr(original_dataset, "get_labels") and hasattr(original_dataset, "get_classes"):
            all_labels = original_dataset.get_labels()
            self._all_classes = original_dataset.get_classes()
            self._all_class_to_idx = original_dataset.get_class_to_idx()

            self._subset_labels = [all_labels[i] for i in self.indices]
            self._subset_labels = [int(lbl) for lbl in self._subset_labels]
            unique_labels_in_subset = sorted(list(set(self._subset_labels)))

            self._subset_classes = [self._all_classes[i] for i in unique_labels_in_subset]
            self._subset_class_to_idx = {
                cls_name: i for i, cls_name in enumerate(self._subset_classes)
            }
        else:
            # fallback
            self._subset_labels = []
            labels = []
            for i in tqdm(self.indices, desc="Inferring labels for subset"):
                _, lab = original_dataset[i]
                labels.append(lab)
            self._subset_labels = [int(lab) for lab in labels]
            unique_labels_in_subset = sorted(list(set(self._subset_labels)))
            self._subset_classes = [f"Class_{lbl}" for lbl in unique_labels_in_subset]
            self._subset_class_to_idx = {nm: i for i, nm in enumerate(self._subset_classes)}

    def __len__(self):
        return len(self.indices)

    def __getitem__(self, idx):
        return self.subset[idx]

    def get_labels(self):
        return self._subset_labels


class SubsetWithTransform(Dataset):
    def __init__(self, subset: SubsetWithLabels, transform=None):
        self.subset = subset
        self.transform = transform

    def __len__(self):
        return len(self.subset)

    def __getitem__(self, idx):
        image, label = self.subset[idx]
        if self.transform:
            image = self.transform(image)
        return image, label

    def get_labels(self):
        return self.subset.get_labels()

###############################################################################
# 3) Hyperparameters
###############################################################################
N_WAY = 5
N_SHOT = 5
N_QUERY = 15
N_VAL_EPISODES = 1000
N_EVALUATION_TASKS = 1000

N_TRAINING_EPOCHS = 25
BATCH_SIZE_FT = 32
LEARNING_RATE_FT = 1e-4
WEIGHT_DECAY_FT = 1e-5
PATIENCE_FT = 5

DISTILL_ALPHA = 1.0
TRIPLET_BETA  = 1.0
TRIPLET_MARGIN= 1.0
DISTILL_EPOCHS= 50
DISTILL_LR    = 1e-4
DISTILL_PATIENCE= 5
DISTILL_EVAL_FREQ= 5

P_CLASSES=8
K_SAMPLES=4
image_size=224

root_dir= "/content/drive/MyDrive/WeedClassification/Datasets/WeedNet"
model_save_dir= "/content/drive/MyDrive/WeedClassification/saved_models/"
os.makedirs(model_save_dir, exist_ok=True)

###############################################################################
# 4) Transforms
###############################################################################
train_transform= transforms.Compose([
    transforms.RandomResizedCrop(image_size, scale=(0.8,1.0)),
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomVerticalFlip(p=0.5),
    transforms.RandomRotation(20),
    transforms.ColorJitter(brightness=0.3, contrast=0.3, saturation=0.2, hue=0.1),
    transforms.ToTensor(),
    transforms.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225])
])
val_transform= transforms.Compose([
    transforms.Resize(int(image_size*1.15)),
    transforms.CenterCrop(image_size),
    transforms.ToTensor(),
    transforms.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225])
])
test_transform= val_transform

###############################################################################
# 5) Load + Split dataset
###############################################################################
full_ds= WeedNet(root_dir, transform=None)
all_labels= full_ds.get_labels()
num_classes_total= len(set(all_labels))

from sklearn.model_selection import StratifiedShuffleSplit
splitter= StratifiedShuffleSplit(n_splits=1, test_size=0.30, random_state=seed)
train_idx, valtest_idx= next(splitter.split(range(len(full_ds)), all_labels))

valtest_labels= [all_labels[i] for i in valtest_idx]
splitter2= StratifiedShuffleSplit(n_splits=1, test_size=0.5, random_state=seed)
val_idx_part, test_idx_part= next(splitter2.split(valtest_idx, valtest_labels))

train_sub= Subset(full_ds, train_idx)
val_sub  = Subset(full_ds, [valtest_idx[i] for i in val_idx_part])
test_sub = Subset(full_ds, [valtest_idx[i] for i in test_idx_part])

train_labeled= SubsetWithLabels(train_sub)
val_labeled  = SubsetWithLabels(val_sub)
test_labeled = SubsetWithLabels(test_sub)

train_dataset= SubsetWithTransform(train_labeled, transform=train_transform)
val_dataset  = SubsetWithTransform(val_labeled,   transform=val_transform)
test_dataset = SubsetWithTransform(test_labeled,  transform=test_transform)

print(f"Train samples: {len(train_dataset)}, Val samples: {len(val_dataset)}, Test samples: {len(test_dataset)}")
print(f"Number of Classes (Total): {num_classes_total}")

###############################################################################
# 6) Teacher Model (EffNet-B7 partial fine-tune)
###############################################################################
from torchvision.models import efficientnet_b7, EfficientNet_B7_Weights

teacher_model= efficientnet_b7(weights=EfficientNet_B7_Weights.IMAGENET1K_V1)
teacher_model.classifier= nn.Sequential(
    nn.Dropout(p=0.4, inplace=True),
    nn.Linear(in_features=2560, out_features=2048)
)
teacher_num_classes= num_classes_total
teacher_class_layer= nn.Linear(2048, teacher_num_classes)

for name, param in teacher_model.named_parameters():
    if "features.6" in name or "features.7" in name or "features.8" in name or "classifier" in name:
        param.requires_grad= True
    else:
        param.requires_grad= False

teacher_class_layer.weight.requires_grad= True
teacher_class_layer.bias.requires_grad= True

teacher_model= teacher_model.cuda()
teacher_class_layer= teacher_class_layer.cuda()

try:
    if torch.__version__>= "2.0.0":
        torch._dynamo.config.force_parameter_static_shapes= False
        teacher_model= torch.compile(teacher_model)
except:
    pass

class TeacherClassifier(nn.Module):
    def __init__(self, backbone, classifier):
        super().__init__()
        self.backbone= backbone
        self.classifier= classifier
    def forward(self,x):
        emb= self.backbone(x)
        return self.classifier(emb)

def partial_fine_tune_teacher(
    teacher_backbone,
    teacher_classifier,
    train_data,
    val_data,
    test_data=None,
    epochs=50,
    lr=1e-4,
    patience=5,
    batch_size=32,
    weight_decay=1e-5
):
    combined= TeacherClassifier(teacher_backbone, teacher_classifier).cuda()
    optimizer= optim.Adam(combined.parameters(), lr=lr, weight_decay=weight_decay)
    scheduler= optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=epochs, eta_min=1e-6)
    criterion= nn.CrossEntropyLoss()

    train_loader= DataLoader(train_data, batch_size=batch_size, shuffle=True, num_workers=4)
    val_loader  = DataLoader(val_data,   batch_size=batch_size, shuffle=False,num_workers=4)

    best_val_acc= -1
    patience_cnt=0
    best_model_state= copy.deepcopy(combined.state_dict())

    scaler= torch.cuda.amp.GradScaler()

    for ep in range(epochs):
        combined.train()
        run_loss=0.0
        correct_train=0
        total_train=0
        for imgs,lbls in train_loader:
            imgs,lbls= imgs.cuda(), lbls.cuda()
            optimizer.zero_grad()
            with torch.cuda.amp.autocast():
                outs= combined(imgs)
                loss= criterion(outs,lbls)
            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()

            run_loss+= loss.item()
            preds= torch.argmax(outs,dim=1)
            correct_train+= (preds==lbls).sum().item()
            total_train+= lbls.size(0)

        train_loss= run_loss/ len(train_loader)
        train_acc= 100.0* correct_train/ total_train if total_train else 0.0

        # val
        combined.eval()
        correct_val=0
        total_val=0
        with torch.no_grad(), torch.cuda.amp.autocast():
            for vimgs,vlbls in val_loader:
                vimgs,vlbls= vimgs.cuda(), vlbls.cuda()
                vouts= combined(vimgs)
                vpred= torch.argmax(vouts,dim=1)
                correct_val+= (vpred==vlbls).sum().item()
                total_val+= vlbls.size(0)
        val_acc= 100.0* correct_val/ total_val if total_val else 0.0

        print(f"[Teacher FT] Epoch {ep+1}/{epochs}, Loss={train_loss:.4f}, "
              f"TrainAcc={train_acc:.2f}%, ValAcc={val_acc:.2f}%")

        if val_acc> best_val_acc:
            best_val_acc= val_acc
            patience_cnt=0
            best_model_state= copy.deepcopy(combined.state_dict())
        else:
            patience_cnt+=1
            if patience_cnt>= patience:
                print(f"Early stopping at epoch={ep+1}")
                break
        scheduler.step()

    combined.load_state_dict(best_model_state)
    print(f"[Teacher FT] BestValAcc={best_val_acc:.2f}%")

    if test_data is not None:
        test_loader= DataLoader(test_data,batch_size=batch_size,shuffle=False,num_workers=4)
        combined.eval()
        correct_test=0
        total_test=0
        with torch.no_grad(), torch.cuda.amp.autocast():
            for timgs,tlbls in test_loader:
                timgs,tlbls= timgs.cuda(), tlbls.cuda()
                tout=combined(timgs)
                tpred= torch.argmax(tout,dim=1)
                correct_test+=(tpred==tlbls).sum().item()
                total_test+= tlbls.size(0)
        test_acc= 100.0* correct_test/ total_test if total_test else 0.0
        print(f"[Teacher FT] TestAcc={test_acc:.2f}%")

    final_backbone= copy.deepcopy(combined.backbone)
    return final_backbone

teacher_embedding_extractor= partial_fine_tune_teacher(
    teacher_model,
    teacher_class_layer,
    train_dataset,
    val_dataset,
    test_data=test_dataset,
    epochs=50,
    lr=LEARNING_RATE_FT,
    patience=PATIENCE_FT,
    batch_size=BATCH_SIZE_FT,
    weight_decay=WEIGHT_DECAY_FT
)
###############################################################################
# 7) Single MobileNetV3 Student
###############################################################################
class StudentEmbeddingNet(nn.Module):
    """
    Single student => produce 2048-D embeddings
    """
    def __init__(self, backbone, in_features, dropout_p=0.5):
        super().__init__()
        self.backbone= backbone
        self.pool= nn.AdaptiveAvgPool2d((1,1))
        self.flatten= nn.Flatten()
        self.dropout= nn.Dropout(p=dropout_p)
        self.project= nn.Linear(in_features, 2048)  # same dimension as teacher

    def forward(self,x):
        x= self.backbone(x)
        x= self.pool(x)
        x= self.flatten(x)
        x= self.dropout(x)
        x= self.project(x)
        return x

# Build single MobileNetV3-Small as the student
from torchvision.models import mobilenet_v3_small, MobileNet_V3_Small_Weights
mobv3= mobilenet_v3_small(weights=MobileNet_V3_Small_Weights.IMAGENET1K_V1)
student_single= StudentEmbeddingNet(
    backbone= mobv3.features,
    in_features=576,
    dropout_p=0.5
).cuda()

try:
    if torch.__version__>= "2.0.0":
        torch._dynamo.config.force_parameter_static_shapes= False
        student_single= torch.compile(student_single)
except:
    pass

class SiameseTripletNetwork(nn.Module):
    def __init__(self, backbone):
        super().__init__()
        self.backbone= backbone
    def forward(self,x):
        return self.backbone(x)

fewshot_model= SiameseTripletNetwork(student_single).cuda()

###############################################################################
# 8) Few-shot & Distillation + Triplet
###############################################################################
def evaluate_few_shot_prototypical(model_, dataset_, n_way, n_shot, n_query, n_tasks):
    """
    Standard Prototypical Few-Shot evaluation
    """
    if not easyfsl_available:
        print("easyfsl not installed => skipping few-shot.")
        return 0.0
    model_.eval()
    sampler= TaskSampler(dataset_, n_way=n_way, n_shot=n_shot, n_query=n_query, n_tasks=n_tasks)
    loader= DataLoader(
        dataset_,
        batch_sampler= sampler,
        collate_fn= sampler.episodic_collate_fn,
        num_workers=4
    )
    total_correct=0
    total_count=0

    with torch.no_grad():
        for (support_imgs, support_lbls, query_imgs, query_lbls, _) in loader:
            support_imgs= support_imgs.cuda()
            query_imgs= query_imgs.cuda()
            support_lbls= support_lbls.cpu()
            query_lbls= query_lbls.cpu()

            s_emb= model_(support_imgs).cpu()
            q_emb= model_(query_imgs).cpu()

            unique_labels= torch.unique(support_lbls)
            centroids=[]
            for ulab in unique_labels:
                idx= (support_lbls==ulab)
                centroids.append(s_emb[idx].mean(dim=0))
            centroids= torch.stack(centroids, dim=0)

            dists= torch.cdist(q_emb, centroids)
            preds_idx= torch.argmin(dists, dim=1)
            mapped_preds= [unique_labels[p].item() for p in preds_idx]

            correct=0
            for i, mp in enumerate(mapped_preds):
                if mp== query_lbls[i].item():
                    correct+=1
            total_correct+= correct
            total_count+= len(query_lbls)

    acc= 100.0* total_correct/ total_count if total_count else 0.0
    return acc

def train_combined_distillation(
    model_,
    teacher_extractor,
    train_data,
    val_data,
    p_classes,
    k_samples,
    alpha,
    beta,
    margin,
    epochs=30,
    lr=1e-4,
    weight_decay=1e-5,
    patience=5,
    eval_freq=5,
    test_data=None
):
    if not metric_learning_available or not easyfsl_available:
        print("Metric learning libs not available => skipping training.")
        return [], [], [], [], [], []

    train_loader= DataLoader(train_data, batch_size=64, shuffle=True, num_workers=4)
    criterion_mse= nn.MSELoss()

    from pytorch_metric_learning import miners, losses
    semi_hard_miner= miners.TripletMarginMiner(margin=margin, type_of_triplets="semihard")
    triplet_loss_obj= losses.TripletMarginLoss(margin=margin)

    optimizer= optim.Adam(model_.parameters(), lr=lr, weight_decay=weight_decay)
    scheduler= optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=epochs, eta_min=1e-6)
    scaler= torch.cuda.amp.GradScaler()

    best_val_acc= -1.0
    patience_cnt= 0
    best_model_state= copy.deepcopy(model_.state_dict())

    train_total_loss_hist= []
    train_mse_loss_hist= []
    train_triplet_loss_hist= []
    val_acc_hist= []
    val_loss_hist= []
    eval_epoch_list= []
    train_few_shot_acc_hist= []

    print(f"Starting Distillation+Triplet => alpha={alpha}, beta={beta}, margin={margin}.")

    for ep in range(1, epochs+1):
        model_.train()
        teacher_extractor.eval()

        run_loss_combined= 0.0
        run_loss_mse= 0.0
        run_loss_trip= 0.0
        nbatches= 0

        for imgs, lbls in tqdm(train_loader, desc=f"Epoch {ep}/{epochs} [semi-hard]"):
            imgs, lbls= imgs.cuda(), lbls.cuda()
            with torch.no_grad():
                teacher_embs= teacher_extractor(imgs).detach()
            student_embs= model_(imgs)

            # Distillation MSE
            loss_mse= criterion_mse(student_embs, teacher_embs)

            # Semi-hard triplet
            emb= model_(imgs)  # fresh forward
            triplets= semi_hard_miner(emb, lbls)
            if len(triplets[0])>0:
                loss_trip= triplet_loss_obj(emb, lbls, triplets)
            else:
                loss_trip= torch.tensor(0.0, device=imgs.device)

            total_loss= alpha*loss_mse + beta*loss_trip

            optimizer.zero_grad()
            with torch.cuda.amp.autocast():
                pass
            scaler.scale(total_loss).backward()
            scaler.step(optimizer)
            scaler.update()

            run_loss_combined+= total_loss.item()
            run_loss_mse+= loss_mse.item()
            run_loss_trip+= loss_trip.item()
            nbatches+=1

        ep_combined= run_loss_combined/ nbatches if nbatches else 0.0
        ep_mse= run_loss_mse/ nbatches if nbatches else 0.0
        ep_trip= run_loss_trip/ nbatches if nbatches else 0.0

        train_total_loss_hist.append(ep_combined)
        train_mse_loss_hist.append(ep_mse)
        train_triplet_loss_hist.append(ep_trip)

        scheduler.step()

        # Evaluate few-shot accuracy on train set for logging
        model_.eval()
        train_fs_acc=0.0
        if ep%eval_freq==0 or ep==epochs:
            train_fs_acc= evaluate_few_shot_prototypical(model_, train_data, N_WAY, N_SHOT, N_QUERY, N_VAL_EPISODES)
        train_few_shot_acc_hist.append(train_fs_acc)

        # Evaluate on val set few-shot
        val_fs_acc=0.0
        if ep%eval_freq==0 or ep==epochs:
            print(f"[Distill+Triplet] EVAL epoch={ep}, semi-hard miner")
            if val_data and len(val_data)>0:
                val_fs_acc= evaluate_few_shot_prototypical(model_, val_data, N_WAY, N_SHOT, N_QUERY, N_VAL_EPISODES)
            else:
                print("No val_data => skipping.")
            val_acc_hist.append(val_fs_acc)
            eval_epoch_list.append(ep)

            val_loss_hist.append(0.0)  # skipping separate val loss
            print(f"[Epoch {ep}] CombinedLoss={ep_combined:.4f}, TrainFS={train_fs_acc:.2f}%, ValFS={val_fs_acc:.2f}%")

            # Early stopping on val FS
            if val_fs_acc> best_val_acc:
                best_val_acc= val_fs_acc
                patience_cnt=0
                best_model_state= copy.deepcopy(model_.state_dict())
                print(f"  New best val FS acc={best_val_acc:.2f}%")
            else:
                patience_cnt+=1
                print(f"  No improvement => patience={patience_cnt}/{patience}")
                if patience_cnt>= patience:
                    print(f"Early stopping at epoch={ep}")
                    break

    model_.load_state_dict(best_model_state)
    print(f"Loaded best model => val FS acc={best_val_acc:.2f}%")

    if test_data is not None:
        test_fs_acc= evaluate_few_shot_prototypical(model_, test_data, N_WAY, N_SHOT, N_QUERY, N_EVALUATION_TASKS)
        print(f"[Distill+Triplet] Final Test FS acc={test_fs_acc:.2f}%")

    # Plot the training curves
    if train_total_loss_hist:
        plt.figure(figsize=(10,6))
        plt.plot(range(1,len(train_total_loss_hist)+1), train_total_loss_hist, label="Total Combined Loss")
        plt.plot(range(1,len(train_mse_loss_hist)+1), train_mse_loss_hist, label="MSE Distillation", linestyle='--')
        plt.plot(range(1,len(train_triplet_loss_hist)+1), train_triplet_loss_hist, label="Triplet", linestyle=':')
        plt.title("Distillation + Semi-hard Triplet => Train Loss Over Epochs")
        plt.xlabel("Epoch")
        plt.ylabel("Loss")
        plt.legend()
        plt.grid(True)
        plt.show()

    if train_few_shot_acc_hist and val_acc_hist and eval_epoch_list:
        plt.figure(figsize=(10,6))
        offset= len(train_few_shot_acc_hist)- len(eval_epoch_list)
        train_curve= train_few_shot_acc_hist[offset:] if offset>=0 else train_few_shot_acc_hist
        plt.plot(eval_epoch_list, train_curve, 'b-', label="Train FS Acc")
        plt.plot(eval_epoch_list, val_acc_hist, 'ro-', label="Val FS Acc")
        plt.title("Train & Val Few-Shot Accuracy Over Epochs (Distill+Triplet)")
        plt.xlabel("Epoch")
        plt.ylabel("Accuracy (%)")
        plt.legend()
        plt.grid(True)
        plt.show()

    return (train_total_loss_hist, train_mse_loss_hist, train_triplet_loss_hist,
            val_acc_hist, val_loss_hist, eval_epoch_list)

###############################################################################
# 9) Training with Distillation + Triplet
###############################################################################
if metric_learning_available and easyfsl_available:
    (train_loss_hist,
     train_mse_hist,
     train_triplet_hist,
     val_acc_hist,
     val_loss_hist,
     eval_epochs)= train_combined_distillation(
        fewshot_model,
        teacher_embedding_extractor,
        train_dataset,
        val_dataset,
        p_classes=P_CLASSES,
        k_samples=K_SAMPLES,
        alpha=DISTILL_ALPHA,
        beta=TRIPLET_BETA,
        margin=TRIPLET_MARGIN,
        epochs=DISTILL_EPOCHS,
        lr=DISTILL_LR,
        weight_decay=1e-5,
        patience=DISTILL_PATIENCE,
        eval_freq=DISTILL_EVAL_FREQ,
        test_data=test_dataset
    )
else:
    print("Skipping combined training => metric/easyfsl not installed or missing.")

###############################################################################
# 10) Final N-Way N-Shot (with n_runs=3)
###############################################################################
def multi_run_few_shot_eval(
    model_,
    dataset_,
    ways_list,
    shots_list,
    n_query,
    n_tasks,
    n_runs=3
):
    """
    Evaluate N-way, N-shot multiple times => compute mean ± std
    """
    if not easyfsl_available:
        return

    results= []
    for ways in ways_list:
        for shots in shots_list:
            print(f"\n=== Evaluate: {ways}-Way {shots}-Shot ===")
            run_accuracies= []
            for r in range(n_runs):
                print(f"   Run {r+1}/{n_runs} ...")
                acc= evaluate_few_shot_prototypical(model_, dataset_, ways, shots, n_query, n_tasks)
                run_accuracies.append(acc)
                print(f"   Accuracy: {acc:.2f}%")

            mean_acc= np.mean(run_accuracies)
            std_acc= np.std(run_accuracies)
            print(f"--> {ways}-Way {shots}-Shot => mean={mean_acc:.2f}%, std={std_acc:.2f}%")
            results.append((ways, shots, mean_acc, std_acc))
    return results

if metric_learning_available and easyfsl_available:
    ways_to_try= [3, 5, 7]
    shots_to_try= [1, 5, 10]
    final_results= multi_run_few_shot_eval(
        fewshot_model,
        test_dataset,
        ways_list= ways_to_try,
        shots_list= shots_to_try,
        n_query= N_QUERY,
        n_tasks= N_EVALUATION_TASKS,
        n_runs= 3   # <-- changed from 5 to 3
    )
    print("\n----- Final (N-Way, N-Shot) Results on Test -----")
    for (w, s, m, sd) in final_results:
        print(f"{w}-Way {s}-Shot => Mean={m:.2f}% +/- {sd:.2f}%")
else:
    print("Skipping final N-Way => easyfsl not available.")

###############################################################################
# 11) Quantize Model for Edge
###############################################################################
if metric_learning_available and easyfsl_available:
    save_path= os.path.join(model_save_dir, "fewshot_weednet_mobilenetv3_single.pth")
    torch.save(fewshot_model.state_dict(), save_path)
    orig_size= os.path.getsize(save_path)
    print(f"Original model size: {orig_size/1e6:.2f} MB")

    # Try dynamic quant
    quant_model= copy.deepcopy(fewshot_model).cpu()
    quant_model.eval()
    try:
        quant_model= torch.quantization.quantize_dynamic(
            quant_model, {nn.Linear}, dtype=torch.qint8
        )
        quant_path= os.path.join(model_save_dir, "fewshot_weednet_mobilenetv3_single_quant.pth")
        torch.save(quant_model.state_dict(), quant_path)
        quant_size= os.path.getsize(quant_path)
        red= 100.0*(orig_size- quant_size)/ orig_size
        print(f"Quantized model size: {quant_size/1e6:.2f} MB => reduced by {red:.2f}%")
    except Exception as e:
        print(f"Quantization failed: {e}")
else:
    print("Skipping quant => missing easyfsl or training not done.")

###############################################################################
# 12) Inference Speed
###############################################################################
def measure_few_shot_inference_speed(model_, device='cuda', n_way=5, n_shot=5, n_query=5, num_tasks=10):
    """
    Timed run of prototypical logic
    """
    if not easyfsl_available:
        print("easyfsl not installed => skipping.")
        return
    is_quant= any("Quantized" in str(type(m)) for m in model_.modules())
    if is_quant:
        device='cpu'
        print("Quantized => forcing CPU.")
    model_.eval().to(device)

    support= torch.randn(n_way*n_shot,3,image_size,image_size, device=device)
    query= torch.randn(n_way*n_query,3,image_size,image_size, device=device)
    support_labels= torch.arange(n_way).unsqueeze(1).expand(n_way,n_shot).reshape(-1).to(device)

    # warmup
    for _ in range(2):
        with torch.no_grad():
            s_emb= model_(support)
            q_emb= model_(query)
            cents=[]
            for w in range(n_way):
                idx= (support_labels== w)
                cents.append(s_emb[idx].mean(dim=0))
            cents= torch.stack(cents, dim=0)
            dists= torch.cdist(q_emb, cents)

    if device=='cuda':
        torch.cuda.synchronize()
    t0= time.time()

    for _ in range(num_tasks):
        with torch.no_grad():
            s_emb= model_(support)
            q_emb= model_(query)
            cents=[]
            for w in range(n_way):
                idx= (support_labels==w)
                cents.append(s_emb[idx].mean(dim=0))
            if not cents:
                break
            cents= torch.stack(cents, dim=0)
            _= torch.cdist(q_emb, cents)

    if device=='cuda':
        torch.cuda.synchronize()
    t1= time.time()

    total_time= t1- t0
    if num_tasks>0:
        avg_time= total_time/ num_tasks
        total_imgs= (n_way*n_shot)+(n_way*n_query)
        throughput= (total_imgs*num_tasks)/ total_time
        print(f"\n[Few-Shot Inference Speed => {n_way}-Way, {n_shot}-Shot, {n_query}-Query, tasks={num_tasks}]")
        print(f"  average time/task={avg_time:.4f}s, throughput={throughput:.2f} samples/s")

    process= psutil.Process(os.getpid())
    mem_info= process.memory_info().rss/(1024**2)
    print(f"CPU Memory Usage= {mem_info:.2f} MB")

    if gpu_monitor_available and device=='cuda':
        try:
            handle= pynvml.nvmlDeviceGetHandleByIndex(0)
            info= pynvml.nvmlDeviceGetMemoryInfo(handle)
            print(f"GPU Memory Usage= {info.used/(1024**2):.2f} MB")
        except:
            pass

if metric_learning_available and easyfsl_available:
    measure_few_shot_inference_speed(fewshot_model, device='cuda', n_way=5, n_shot=5, n_query=5, num_tasks=10)
else:
    print("Skipping speed test => missing easyfsl.")

print("\n===== Pipeline Complete (Single MobileNetV3 Student) =====\n"
      "1) Teacher (EffNet-B7) partial fine-tune.\n"
      "2) Single MobileNetV3 student => Distillation + Triplet.\n"
      "3) Few-shot evaluations with n_runs=3.\n"
      "4) Optional dynamic quantization.\n"
      "5) Speed test.\n")
```
